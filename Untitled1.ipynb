```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import accuracy_score
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
from tensorflow.keras.models import Sequential
```

```python
(x_train,y_train),(x_test,y_test) = tf.keras.datasets.mnist.load_data()
```

```python
num_pixels = x_train.shape[1] * x_train.shape[2]
df_train = pd.DataFrame(x_train.reshape(-1, num_pixels))
df_train['label'] = y_train
print(df_train.head())
```

```python
pixel_columns = [col for col in df_train.columns if col
!= 'label']
verage_digits = df_train.groupby('label')[pixel_columns].mean().values
```

```python
x_train_flat = x_train.reshape(-1, 28*28)
x_test_flat = x_test.reshape(-1, 28*28)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(x_train_flat, y_train)
```

```python
y_pred = knn.predict(x_test_flat)
```

```python
x_train_norm = x_train / 255.0
x_test_norm = x_test / 255.0
```

```python
y_train_cat = to_categorical(y_train, num_classes=10)
y_test_cat = to_categorical(y_test, num_classes=10)
```

```python
from keras.layers import Dense, Activation, Flatten
```

```python
model = Sequential([
 Flatten(input_shape=(28, 28)),
 Dense(128, activation='relu'),
 Dense(10, activation='softmax')
])
model.compile(optimizer='adam',
loss='categorical_crossentropy', metrics=['accuracy'])
history = model.fit(x_train_norm, y_train_cat, epochs=5,
validation_split=0.1)
```

```python
test_loss, test_acc = model.evaluate(x_test_norm,
y_test_cat)
print('Точность нейронной сети на тестовой выборке:',
test_acc)
```
